#### 注意：颜色通道顺序
1. OpenCV (cv2):
OpenCV 默认使用 BGR（蓝色、绿色、红色）顺序来存储和显示图像。
这意味着在 OpenCV 中读取和处理的图像，其颜色通道顺序是蓝色在前、红色在后。
2. Matplotlib (plt):Matplotlib 使用的颜色顺序是 RGB（红色、绿色、蓝色）。
    这意味着 Matplotlib 显示图像时，红色在前，蓝色在后。

```python
import cv2

image = cv2.imread('500X300.png') #BGR格式
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) #转为RGB格式
image.shape  #（高、宽、颜色通道）
# print(image.shape, image[:5])
# cv2.imshow('image', image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
# 第一维度：前2个元素  第二维度：每行500个值 第三维度：RGB三个值
image[:2].shape
```

```python
plt.imshow(image)
```

```python
# 修改区域颜色
start_row, start_col = 50, 100
end_row, end_col = 250, 300

# 修改该区域的像素值为红色 (RGB: 255, 0, 0)
imageRed = image.copy()
imageRed[start_row:end_row, start_col:end_col] = [255, 0, 0]
plt.imshow(imageRed)
```
